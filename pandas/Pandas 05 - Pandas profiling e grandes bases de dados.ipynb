```python
import pandas_profiling
import pandas as pd
```

## Utilizando o Pandas Profiling

Lendo o arquivo csv utilizando o pandas

```python
arquivo = 'kc_house_data.csv'
df = pd.read_csv(arquivo, sep=',', header=0)
df.head()
```

Utilizando o pandas profiling para inspecionar o dataframe.

```python
pandas_profiling.ProfileReport(df)
```

Salvando o relatório em um arquivo HTML:

```python
profile = pandas_profiling.ProfileReport(df)
profile.to_file(output_file='report.html')
```
