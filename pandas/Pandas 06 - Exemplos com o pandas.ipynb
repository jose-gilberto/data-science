## Removendo linhas que possuem ao menos um registro faltante

## Preenchendo com 1 os valores faltantes da coluna `bedrooms`
*Nesse caso esta função não irá surtir efeito pois os dados faltantes já foram removidos do DataFrame, mas será executada com a finalidade de servir de exemplo.

## Exemplos de consultas com o pandas
### Todos os imóveis que tenham mais de 1 banheiros (`bathroom`)

```python
df[df['bathrooms'] > 1]
```

### Todos os imoveis que tenham a quantidade de quartos (`bedrooms`) igual a 4.

```python
df[df['bedrooms'] == 4]
```

### Contagem do retorno da consulta anterior

```python
df[df['bedrooms'] == 4].shape
```

### Ordene o DataFrame pelos preços (`price`) em ordem decrescente

```python
df.sort_values(by='price', inplace=True)
df.head()
```

### Agrupe e conte quantidade de imóveis por tamanhos

```python
def size_category(square_foot):
    if square_foot < 500:
        return 'small'
    elif (square_foot > 500) & (square_foot < 800):
        return 'medium'
    else:
        return 'large'
```

```python
df['size'] = df['sqft_living']
df.head()
```

```python
df['size'] = df['size'].apply(size_category)
df.head()
```

#### Contando os imóveis de cada tamanho

```python
df['size'].value_counts()
```
